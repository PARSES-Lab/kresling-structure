```python
import numpy as np
import matplotlib.pyplot as plt
import csv

with open('/Users/kris/Desktop/kresling_R3_alpha30_H15_W0_L407_SF_1.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    row_count = 0
    disp = np.array([])
    time = np.array([])
    force = np.array([])
    for row in csv_reader:
        if row_count > 3:
            disp = np.append(disp, float(row[1]))
            time = np.append(time, float(row[2]))
            force = np.append(force, float(row[0]))
        else:
            row_count += 1
            
abs_force = np.absolute(force)           
zero_start = np.where(abs_force == np.amin(abs_force))
print(zero_start)
disp = disp - disp[zero_start][0]
force = force - force[zero_start][0]
time = time - time[zero_start][0]

fig, ax = plt.subplots()
ax.plot(disp,force)

ax.set(xlabel='Time (s)', ylabel='Force (N)')
ax.grid()

fig.savefig("/Users/kris/Desktop/disp_and_force_SF1.png")
plt.show()

```
