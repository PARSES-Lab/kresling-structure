# Kresling Compression Test Analysis
The following code is used to plot the results of a Kresling compression test.

## Import Relevant Modules
Code runs on `numpy`, `matplotlib`, and `csv`. `numpy` is referred to as `np` and `matplotlib.pyplot` is referred to as `plt` for simplicity.

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

## Import .csv
To analyze the results of the test, the .csv file containing the test results must be opened. The `with` statement is a form of exception handling that uses the `__enter__()` and `__exit__()` methods built into `open()` that ensures the following code is run only when the file is opened correctly. It automatically closes the file at the end of the code block. ([Reference](https://www.geeksforgeeks.org/with-statement-in-python/))

The `as` keyword assigns an alias to a resource. In other words, `csv_file` is the alias for (a variable that refers to) the .csv file we're opening. ([Reference](https://www.geeksforgeeks.org/python-as-keyword/))

A csv reader object for our test results is stored in `csv_reader`. This object is essentially a collection of all the rows in our test results file, and every row is a list with a string for each column. ([Reference](https://docs.python.org/3/library/csv.html)) 

Numpy arrays are built to collect the three variables we care about: displacement, time, and force.

Now the lists in the csv reader object are iterated over so we can collect the relevant data in our numpy arrays. `row_count` ensures we skip the first four rows, which consist of empty cells and headers in the test results file. After the first four rows, we append the value from the corresponding column in each of our numpy arrays (ex: the first value \[0] is the first column, which has force values).

```python
with open('kresling_L0.75_FS_A.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    row_count = 0
    disp = np.array([])
    time = np.array([])
    force = np.array([])
    for row in csv_reader:
        if row_count > 3:
            disp = np.append(disp, float(row[1]))
            time = np.append(time, float(row[2]))
            force = np.append(force, float(row[0]))
        else:
            row_count += 1      
```

## Zeroing out the data
Now the three numpy arrays contain all the data for displacement, time, and force from the test results file. We want to shift the values in each of these arrays so that the force values begin at zero, the time values begin at zero when the force is zero, and the displacement values begin at zero when the force is zero.

First, a new `abs_force` array is constructed to store the absolute value of the original force array. ([Reference](https://numpy.org/doc/stable/reference/generated/numpy.absolute.html)) `np.amin()` is called with `abs_force` as an input, which will output the positive force value that is closest to zero. ([Reference](https://numpy.org/doc/stable/reference/generated/numpy.amin.html))

Then, the `np.where()` method is called. When `np.where(condition)` is called, it acts as shorthand for `np.asarray(condition).nonzero()`. ([Reference](https://numpy.org/doc/stable/reference/generated/numpy.where.html)) This means that first `np.asarray(condition)` is called: this outputs an array of boolean values indicating whether the condition is True or False for each element of the compared arrays in the conditional statement. ([Reference](https://numpy.org/doc/stable/reference/generated/numpy.asarray.html)) This array full of boolean values is then sent into the `nonzero()` method, which returns the indices where the array is nonzero OR True. `nonzero()` outputs these indices as a tuple. In other words, if the input was a 2-D array, the output would have two arrays: the first array with the indices indicating the row of each nonzero value, and the second array indicating the column of each nonzero value. ([Reference](https://numpy.org/doc/stable/reference/generated/numpy.nonzero.html#numpy.nonzero))

We're using `np.where()` to output a 1-D array listing the indices of every data point where the absolute value of the force measurements is equal to a value closest to zero. This array of indices is stored in `zero_start`. The first value of this array, `zero_start[0]`, is going to be where the force first equals this value. Now we're going to find the displacement, force, and time values at this indice `zero_start[0]`. These values are subtracted from their respective arrays. 

This shifts our force values so that they begin at zero. This shifts our time values so that t=0 is where the force value is zero. This shifts our displacement values so that disp=0 is where the force value is zero.

**ADD NOTE TO SUBTRACT 135g LATER**

```python
abs_force = np.absolute(force)   
zero_start = np.where(abs_force == np.amin(abs_force))
print(np.asarray(abs_force == np.amin(abs_force)))
print(zero_start)
print(force[zero_start])
disp = disp - disp[zero_start][0]
force = force - force[zero_start][0]
time = time - time[zero_start][0]
```

## Plotting the data
Now the data is ready to be plotted.

```python
fig, ax = plt.subplots()
ax.plot(disp,force)

ax.set(xlabel='Displacement (mm)', ylabel='Force (N)')
ax.grid()

fig.savefig("[kresling_L0.75_FS_A]_force_vs_disp.png")
plt.show()
```
